```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_out_init = pd.read_csv('./_utils/lr_step_testing_init.csv')
df_out_init_ds = pd.read_csv('./_utils/lr_step_testing_downscaled_init.csv')
df_out = pd.read_csv('./_utils/lr_step_testing.csv')
df_out_ds = pd.read_csv('./_utils/lr_step_testing_downscaled.csv')
```

```python
df_out_init = pd.concat([df_out_init, df_out_init_ds])
df_out = pd.concat([df_out, df_out_ds])
```

```python
g = sns.FacetGrid(col='inputs', row='resolution', hue='WD', data=df_out_init, height=3, aspect=1.6, legend_out=True)
g.map_dataframe(sns.lineplot, x='LR', y='train_fscore')
g.map_dataframe(sns.lineplot, x='LR', y='test_fscore', linestyle='dashed')
#g.set_xlim((0.0001, 0.02))
g.set(xscale="log")
#g.set(yscale="log")
g.add_legend()
#plt.axhline(y=0.05, linestyle='dotted', c='black')
plt.show()
```

```python
g = sns.FacetGrid(col='inputs', row='resolution', hue='WD', data=df_out_init, height=3, aspect=1.6, legend_out=True)
g.map_dataframe(sns.lineplot, x='LR', y='train_loss')
g.map_dataframe(sns.lineplot, x='LR', y='test_loss', linestyle='dashed')
#g.set_xlim((0.0001, 0.02))
g.set(xscale="log")
#g.set(yscale="log")
g.add_legend()
#plt.axhline(y=0.05, linestyle='dotted', c='black')
plt.show()
```

```python
subset = True
if subset:
    df_out_init = df_out_init[(df_out_init['inputs'] == 'rgb_tpi_ndvi') &
                              (df_out_init['WD'] == 10e-4)]
    df_out_init['momentum'] = 0.0
```

```python
df_out = pd.concat([df_out, df_out_init])
```

```python
g = sns.FacetGrid(col='inputs', row='resolution', hue='momentum', data=df_out, height=3, aspect=1.6, legend_out=True)
g.map_dataframe(sns.lineplot, x='LR', y='train_fscore')
g.map_dataframe(sns.lineplot, x='LR', y='test_fscore', linestyle='dashed')
#g.set_xlim((0.0001, 0.02))
g.set(xscale="log")
#g.set(yscale="log")
g.add_legend()
#plt.axhline(y=0.05, linestyle='dotted', c='black')
plt.show()
```
